### **Real Time Website Scraping**

```
import requests
```

```
html_text=requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=").text
```

```
soup1=BeautifulSoup(html_text,'lxml')
jobs=soup1.find_all('li',class_='clearfix job-bx wht-shd-bx')
print('Enter the skill that you are not familiar with')
unfamiliar_skill=input('>')
```

```
def find_jobs(file_num):
  with open(f'/content/drive/My Drive/posts/{file_num}.txt','w') as f:
    for job in jobs:
      pub_date=job.find('span',class_='sim-posted').span.text
      #print(pub_date)
      if 'few' in pub_date:
        comp_name=job.h3.text.replace(' ','')
        More_info=job.header.h2.a['href']
        skills=job.find('span',class_='srp-skills').text.replace(' ','')
        if unfamiliar_skill not in skills:
          f.write(f'company Name    : {comp_name.strip()}\n')
          f.write(f'Required_skills : {skills.strip()}\n')
          f.write(f'More_info       : {More_info}\n')
          f.write('')
          file_num+=1
```

```
file_num=1
import time
while True:
  find_jobs(file_num)
  time_wait=1
  print(f'waiting {time_wait} minutes.....')
  time.sleep(time_wait*60)
  file_num+=1
```
